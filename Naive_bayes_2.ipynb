Q2. What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

Difference between Bernoulli Naive Bayes and Multinomial Naive Bayes:

Bernoulli Naive Bayes: Suitable for binary feature data. It assumes that features are binary variables, and it models the presence or absence of a particular feature.

Multinomial Naive Bayes: Appropriate for discrete feature data. It is commonly used in text classification, where features represent word counts or term frequencies.

Q3. How does Bernoulli Naive Bayes handle missing values?

In Bernoulli Naive Bayes, missing values are typically treated as the absence of the feature. If a feature is missing, it is considered as not occurring rather than assigning a specific value.

Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Yes, Gaussian Naive Bayes can be used for multi-class classification. In the context of multi-class classification, each class is modeled as a Gaussian distribution with its own mean and variance for each feature. The class with the highest probability for a given set of features is then predicted.

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the Spambase dataset
data_path = 'path/to/spambase.data'
column_names = [
    # List of column names based on the dataset description
]
df = pd.read_csv(data_path, header=None, names=column_names)

# Assuming the last column is the target variable
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Implement Bernoulli Naive Bayes
bernoulli_nb = BernoulliNB()
accuracy_scores_b = cross_val_score(bernoulli_nb, X, y, cv=10, scoring='accuracy').mean()
precision_scores_b = cross_val_score(bernoulli_nb, X, y, cv=10, scoring='precision').mean()
recall_scores_b = cross_val_score(bernoulli_nb, X, y, cv=10, scoring='recall').mean()
f1_scores_b = cross_val_score(bernoulli_nb, X, y, cv=10, scoring='f1').mean()

# Implement Multinomial Naive Bayes
multinomial_nb = MultinomialNB()
accuracy_scores_m = cross_val_score(multinomial_nb, X, y, cv=10, scoring='accuracy').mean()
precision_scores_m = cross_val_score(multinomial_nb, X, y, cv=10, scoring='precision').mean()
recall_scores_m = cross_val_score(multinomial_nb, X, y, cv=10, scoring='recall').mean()
f1_scores_m = cross_val_score(multinomial_nb, X, y, cv=10, scoring='f1').mean()

# Implement Gaussian Naive Bayes
gaussian_nb = GaussianNB()
accuracy_scores_g = cross_val_score(gaussian_nb, X, y, cv=10, scoring='accuracy').mean()
precision_scores_g = cross_val_score(gaussian_nb, X, y, cv=10, scoring='precision').mean()
recall_scores_g = cross_val_score(gaussian_nb, X, y, cv=10, scoring='recall').mean()
f1_scores_g = cross_val_score(gaussian_nb, X, y, cv=10, scoring='f1').mean()

# Display results
print("Bernoulli Naive Bayes:")
print("Accuracy:", accuracy_scores_b)
print("Precision:", precision_scores_b)
print("Recall:", recall_scores_b)
print("F1 Score:", f1_scores_b)
print()

print("Multinomial Naive Bayes:")
print("Accuracy:", accuracy_scores_m)
print("Precision:", precision_scores_m)
print("Recall:", recall_scores_m)
print("F1 Score:", f1_scores_m)
print()

print("Gaussian Naive Bayes:")
print("Accuracy:", accuracy_scores_g)
print("Precision:", precision_scores_g)
print("Recall:", recall_scores_g)
print("F1 Score:", f1_scores_g)
```
