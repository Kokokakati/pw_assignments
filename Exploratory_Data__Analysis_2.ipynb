```python
#1

import pandas as pd

# Load the dataset
flight_data = pd.read_csv('flight_price_dataset.csv')

# Examine dimensions of the dataset
print("Number of rows:", flight_data.shape[0])
print("Number of columns:", flight_data.shape[1])
```

```python
#2
import matplotlib.pyplot as plt

# Create histogram
plt.hist(flight_data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#3
# Calculate range of prices
price_min = flight_data['Price'].min()
price_max = flight_data['Price'].max()

print("Minimum Price:", price_min)
print("Maximum Price:", price_max)
print("Price Range:", price_max - price_min)
```

```python
#4
import seaborn as sns

# Create boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=flight_data)
plt.title('Flight Prices by Airline')
plt.xticks(rotation=45)
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()
```

```python
#5
# Identify outliers using boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(flight_data['Price'])
plt.title('Identifying Outliers in Flight Prices')
plt.xlabel('Price')
plt.show()
```

Potential Impact: Outliers may skew the mean and standard deviation of the dataset, leading to biased analysis. They may also affect the performance of certain machine learning algorithms that are sensitive to outliers.
