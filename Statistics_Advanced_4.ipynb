Q1: The main difference between a t-test and a z-test lies in the context of their application and the assumptions they make:

T-test: It is used when you have a small sample size (typically less than 30) or when you don't know the population standard deviation. For example, you might use a t-test to compare the means of two groups of students (Group A and Group B) who were taught by different teachers.

Z-test: It is used when you have a large sample size (typically more than 30) and you know the population standard deviation. For instance, if you want to determine if the average height of a population is different from a known national average, you would use a z-test.


Q2:

One-tailed test: In a one-tailed test, you are interested in testing a specific direction of an effect, either greater than or less than a certain value. For example, testing if a new drug increases patient recovery time (greater than a current treatment) would be a one-tailed test.

Two-tailed test: In a two-tailed test, you are interested in testing whether there is an effect in any direction, greater or less than a certain value. For instance, testing if a new drug has any effect on patient recovery time would be a two-tailed test.


Q3:

Type 1 error: This occurs when you incorrectly reject a true null hypothesis. For example, in a criminal trial, convicting an innocent person is a Type 1 error.

Type 2 error: This occurs when you fail to reject a false null hypothesis. In a medical test, a Type 2 error would be declaring a healthy person as sick.




Q4: Bayes's theorem is a probability theory formula used to update the probability for a hypothesis based on new evidence. It is expressed as:

P(A∣B)= P(B∣A)⋅P(A)/P(B)
​

Where:

P(A∣B) is the updated probability of hypothesis A given evidence B.

P(B∣A) is the probability of evidence B given hypothesis A.

P(A) is the prior probability of hypothesis A.

P(B) is the probability of evidence B.

Example: Suppose you want to find the probability of someone having a disease (A) given that they tested positive (B). You have the prior probability of the disease in the population (P(A)), the accuracy of the test (P(B|A)), and the overall probability of testing positive (P(B)). Bayes's theorem helps you update the probability of having the disease based on the test result and prior knowledge.

Q8: The margin of error in a confidence interval represents the range around the sample statistic (e.g., mean or proportion) within which you are reasonably confident the true population parameter lies. The margin of error is affected by the sample size; as the sample size increases, the margin of error decreases.

Example: Suppose you want to estimate the mean income of a city with a 95% confidence level. If you survey 1000 people, your margin of error might be $100. If you survey 2000 people, your margin of error might decrease to $70. A larger sample size provides more information and reduces uncertainty.
