Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

loc: It is used to access a group of rows and columns by labels or a boolean array. It means you can access rows and columns by their names. It takes row and column labels as inputs.
iloc: It is used to access a group of rows and columns by integer position. It means you can access rows and columns by their integer index positions. It takes integer-based row and column indices as inputs.

Q6. What do you understand about the windows function in pandas and list the types of windows functions?

In pandas, the "window" function is used for performing calculations on a sliding window of data in a DataFrame or Series. The window function is powerful for analyzing and processing data over a specific window of rows, where you can define the size of the window and apply various aggregate functions to the data within that window.

Types of window functions in pandas:

Rolling Window: It defines a fixed-size window and calculates statistics over that window as it moves through the data. This is typically used for time-series data or any sequential data.
Expanding Window: It starts with a minimum window size and gradually expands to include more data points as it moves through the data. It is used to compute cumulative statistics.
Exponentially Weighted Moving Average (EWMA): It calculates weighted averages where the weights decrease exponentially with the data's age.

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas timedelta. The program should prompt the user to enter the dates and display the result. Did you observe any difference in both the outputs? If so then explain it.

```python
import pandas as pd

def calculate_time_difference():
    date1_str = input("Enter the first date (YYYY-MM-DD): ")
    date2_str = input("Enter the second date (YYYY-MM-DD): ")

    date_format = '%Y-%m-%d'
    date1 = pd.to_datetime(date1_str, format=date_format)
    date2 = pd.to_datetime(date2_str, format=date_format)

    time_difference = date2 - date1
    days_difference = time_difference.days
    hours_difference = time_difference.seconds // 3600
    minutes_difference = (time_difference.seconds // 60) % 60

    print("Time Difference:")
    print(f"{days_difference} days, {hours_difference} hours, and {minutes_difference} minutes.")

calculate_time_difference()
```

Explanation:
When calculating the time difference between two dates, using Pandas timedelta ensures that the result is accurate, even when the dates span across different months and years. Pandas handles leap years and varying month lengths properly, leading to accurate time differences in days, hours, and minutes.

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
import pandas as pd

def convert_column_to_categorical():
    # Get the file path, column name, and category order from the user
    file_path = input("Enter the CSV file path: ")
    column_name = input("Enter the column name to convert to categorical: ")
    category_order = input("Enter the category order (comma-separated values): ").split(",")

    # Read the CSV file
    df = pd.read_csv(file_path)

    # Convert the specified column to categorical with the given category order
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

    # Sort the data by the categorical column
    sorted_df = df.sort_values(by=column_name)

    # Display the sorted data
    print(sorted_df)

convert_column_to_categorical()
```

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data():
    # Get the file path from the user
    file_path = input("Enter the CSV file path: ")

    # Read the CSV file
    df = pd.read_csv(file_path)

    # Assuming the CSV file has columns 'Product', 'Date', and 'Sales'
    # If the date column is not in datetime format, convert it to datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Group the data by 'Product' and 'Date', and calculate the total sales for each product on each date
    sales_data = df.groupby(['Product', 'Date'])['Sales'].sum().unstack()

    # Create a stacked bar chart to visualize the sales data
    sales_data.plot(kind='bar', stacked=True, figsize=(10, 6))

    # Customize the plot
    plt.title("Sales Data by Product Over Time")
    plt.xlabel("Date")
    plt.ylabel("Sales")
    plt.xticks(rotation=45, ha='right')

    # Display the chart
    plt.tight_layout()
    plt.show()

visualize_sales_data()
```

The program should calculate the mean, median, and mode of the test scores and display the results
in a table.

```python
import pandas as pd
from tabulate import tabulate

def calculate_statistics(file_path):
    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(file_path)

    # Calculate the mean, median, and mode of the test scores
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_scores = df['Test Score'].mode().tolist()

    # Display the results in a table
    table = [
        ['Mean', mean_score],
        ['Median', median_score],
        ['Mode', ', '.join(map(str, mode_scores))]
    ]

    headers = ['Statistic', 'Value']

    print(tabulate(table, headers, tablefmt="grid"))

def main():
    file_path = input("Enter the file path of the CSV file containing the student data: ")
    calculate_statistics(file_path)

if __name__ == "__main__":
    main()
```
